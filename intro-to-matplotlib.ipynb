# Intro to matplotlib

### Fast Graphs
Code to make graphs using matplotlib in a fast way

```python
# Importing libraries
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Define arrays
year = [2010, 2012, 2014, 2016, 2018]
revenue = [20000, 50000, 60000, 75000, 85000]
# Plot
plt.plot(year, revenue)
plt.xlabel('Year')
plt.ylabel('Revenue in usd')
plt.title('Revenue')
plt.show()
```

### 2 Graphs at the time
Code to make 2 graphs at the same time

```python
year = [2010, 2012, 2014, 2016, 2018]
revenue = [22000, 55000, 62000, 75000, 85000]
costs = [30000, 32000, 40000, 50000, 60000]
plt.plot(year, revenue) 
plt.plot(year, costs)
plt.xlabel('Year')
plt.ylabel('Money in usd')
plt.title('Revenue vs Costs in awesome inc')
plt.legend(['Revenue', 'Costs'])
plt.show()
```

### Color and type of line
In this case we change the type of line and color for the graph

```python
year = [2010, 2012, 2014, 2016, 2018]
revenue = [22000, 55000, 62000, 75000, 85000]
costs = [30000, 32000, 40000, 50000, 60000]
plt.plot(year, revenue, color='green', linestyle="--", marker="o")
plt.plot(year, costs, color='red', marker="s")
plt.axis([2010,2018,22000,90000])
plt.xlabel('Year')
plt.ylabel('Money in usd')
plt.title('Revenue vs Costs in awesome inc')
plt.legend(['Revenue', 'Costs'])
plt.show()
```

### Subplots
Subplots are independent graphs at once. 

```python
year = [2010, 2012, 2014, 2016, 2018]
revenue = [22000, 55000, 62000, 75000, 85000]
costs = [30000, 32000, 40000, 50000, 60000]
plt.subplot(2,1,1) 
plt.plot(year, revenue, color='green', linestyle="--", marker="o")
plt.subplot(2,1,2)
plt.plot(year, costs, color='red', marker="s")
plt.subplots_adjust(bottom=0.2, wspace=0.5)
plt.show()
```

### Graphs with lists using numpy

```python
# evenly sampled time at 200ms intervals from 0 to 5
t = np.arange(0., 5., 0.2)
# Plot 3 graphs at the same time
plt.plot(t, t, t, t**2, t, t**3)
plt.show()
```

### Bars for graphs: [Source of money spent on pets in US](https://www.tylervigen.com/view_correlation?id=2956)
To do graphs of bars

```python
year = ["2000", "2001", "2002", "2003", "2004", "2005"]
money =  [8.2, 7, 6.5, 5.3, 5.2, 4]
plt.bar(range(len(year)), money)
#create your ax object here
ax=plt.subplot()
ax.set_xticks(range(len(year)))
ax.set_xticklabels(year, rotation=50)
plt.legend(['Money spent on pets in billions of dollars'])
plt.show()
```

### Pie chart

```python
Topics = ["computer science", "mechanical engineering", "Medicine", "Nursery", "Law"]
Percentage = [35.2, 24.8, 14.5, 15.2, 10.3]
plt.pie(Percentage,autopct='%0.1f%%')
plt.axis('equal')
leg = plt.legend(Topics)
plt.title('Percentage of students in a career')
plt.savefig('students.png') 
# To move the legend box
fig = plt.figure()
plt.draw() # Draw the figure so you can find the positon of the legend.
# Get the bounding box of the original legend
bb = leg.get_bbox_to_anchor().inverse_transformed(ax.transAxes)
# Change to location of the legend. 
xOffset = 0.5
bb.x0 += xOffset
bb.x1 += xOffset
leg.set_bbox_to_anchor(bb, transform = ax.transAxes)
# To plot
plt.show()
```
