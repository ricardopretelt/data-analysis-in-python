# Intro to Matplotlib

## Object oriented interface
Plots are divided into figure and axes objects that are manipulated to get the desired results. A figure object can contain one or more axes objects. The figures are basically the windows, which size and number can be adjusted, and axes control the measures and text.

```python
# Import library
from matplotlib import pyplot as plt
import numpy as np

# Get the data
x = [1, 2, 3, 4, 5]
y = x

# Get the figure object
fig1 = plt.figure()
#fig1 = plt.figure(figsize=(19,9)) #To change the size of the plot window use figsize

# Get the axes object at location in 1x1 grid
ax = fig1.add_subplot(1,1,1)

# Generate the plot
ax.plot(x, y)

# Set labels for x and y axis
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Set title for the plot
ax.set_title('XY plot')

plt.show()
```

```python
# Multiple graphs in the same plot

# Function to get the square of each element in the list
def list_square(a_list):
    return [element**2 for element in a_list]

fig2 = plt.figure()

x = [1, 2, 3, 4, 5]
y = x
z = list_square(x)

# Get the axes instance
ax = fig2.add_subplot(1,1,1)

ax.plot(x, y, label='y')
ax.plot(x, z, label='z')
ax.set_xlabel("X")

# Generate legend
ax.legend()

# Set title
ax.set_title('Two plots')

plt.show()
```

```python
# Two plots in a figure

# Multiple subplots in same figure
fig3 = plt.figure(figsize=(12,9))
x = [1, 2, 3, 4, 5]
y = x
z = list_square(x)

# Divide the figure into 1 row 2 column grid and get the
# axes object for the first column
ax1 = fig3.add_subplot(1,2,1)
ax1.plot(x, y)

# set x and y axis labels
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('y=x plot')

# Get second axes instance in the second column of the 1x2 grid
ax2 = fig3.add_subplot(1,2,2)
ax2.plot(x, z)

ax2.set_xlabel('X')
ax2.set_ylabel('z=X^2')
ax2.set_title('z=x^2 plot')

# Generate the title for the Figure. Note that this is different then the title for individual plots
plt.suptitle("Two plots in a figure")
plt.show()
```

```python
# To create a histogram

mean = 10
std = 0.5
num_samples = 1000
samples = np.random.normal(mean, std, num_samples)

# Get an instance of Figure object
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# Generate histogram plot
ax.hist(samples)
ax.set_xlabel('Sample values')
ax.set_ylabel('Frequency')
ax.set_title('Histogram plot')

plt.show()
```

## Plotting using plt

```python
# Graphs from lists and use labels and change colors
year = [2010, 2012, 2014, 2016, 2018]
revenue = [22000, 55000, 62000, 75000, 85000]
costs = [30000, 32000, 40000, 50000, 60000]
plt.plot(year, revenue, color='green', linestyle="--", marker="o")
plt.plot(year, costs, color='red', marker="s")
plt.axis([2010,2018,22000,90000])
plt.xlabel('Year')
plt.ylabel('Money in usd')
plt.title('Revenue vs Costs in awesome inc')
plt.legend(['Revenue', 'Costs'])
plt.show()
```

```python
# Subplots
year = [2010, 2012, 2014, 2016, 2018]
revenue = [22000, 55000, 62000, 75000, 85000]
costs = [30000, 32000, 40000, 50000, 60000]
plt.subplot(2,1,1) 
plt.plot(year, revenue, color='green', linestyle="--", marker="o")
# To get te axis of a plot in ax
ax=plt.subplot(2,1,2)
plt.plot(year, costs, color='red', marker="s")
plt.subplots_adjust(bottom=0.2, wspace=0.5)
plt.show()
```

```python
# Bar graphs
year = ["2000", "2001", "2002", "2003", "2004", "2005"]
money =  [8.2, 7, 6.5, 5.3, 5.2, 4]
ax=plt.subplot()
plt.bar(range(len(year)), money)
ax.set_xticks(range(len(year)))
ax.set_xticklabels(year, rotation=50)
ax.legend(['Money spent on pets in billions of dollars'])
plt.show()
```

```python
# Side by side bars to compare 2 sets
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 = [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

n1 = 1 # First dataset
n2 = 2 # Second dataset
t = 2 # Number of datasets
d = len(drinks) # Number of sets of bars
w = 0.8 # Width of each bar
store1_x = [t*element + w*n1 for element in range(d)]
plt.bar(store1_x,sales1, label="sales1")
n2=2
store2_x = [t*element + w*n2 for element in range(d)]
plt.bar(store2_x,sales2, label="sales2")
plt.title('Drinks in china and USA', fontsize=20)
plt.xticks(store1_x, drinks)
plt.legend(fontsize=15)
plt.show()
```

```python
# Stacked bars
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]
plt.bar(range(len(drinks)),sales1)
plt.bar(range(len(drinks)),sales2,bottom=sales1)
plt.legend(['Location 1', 'Location 2'])
plt.show()
```

```python
# Bar chart with error
past_years_averages = [82, 84, 83, 86, 74, 84, 90]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006]
error = [1.5, 2.1, 1.2, 3.2, 2.3, 1.7, 2.4]
ax=plt.subplot()
plt.bar(range(len(years)), past_years_averages, yerr=error, capsize= 5)
# Borders for x and y axis
plt.axis([-0.5, 6.5, 70, 95])
ax.set_xticks(range(len(years)))
ax.set_xticklabels(years)
plt.title("Final Exam Averages")
plt.xlabel("Year")
plt.ylabel("Test average")
plt.savefig('images/my_bar_chart.png')
plt.show()
```

## 3D Graphs

```python
# To plot 3d graphs
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.15)
Y = np.arange(-5, 5, 0.15)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis')
```

# Intro to Seaborn

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

## Scatter plot in seaborn

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
# Adding another dimension using a hue semantic (coloring the values 
# with a defined feature)
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips);
```

```python
# Using a palette in the scatter plot
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips);
```

```python
# Using different sizes
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);
```

```python
# For multiple sections
sns.relplot(x="total_bill", y="tip", hue="smoker",
            col="time", data=tips);
```

## Line plots

```python
# Regular line plot
df_18 = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df_18)
g.fig.autofmt_xdate()
```

```python
# Representing uncertainty
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
# Doing a subplot with the hue
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            kind="line", data=fmri);
```

## Categorical scatterplot

```python
# Categorical scatterplot with previous dataset
sns.catplot(x="day", y="total_bill", data=tips);
```

```python
# Categorical scatterplot with hue
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips);
```

## Boxplots

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips);
```

## Violinplots

```python
sns.catplot(x="total_bill", y="day", hue="sex",
            kind="violin", data=tips);
```

## Bar plots

```python
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic);
```

```python
# Bar plot using a palette
sns.catplot(x="deck", kind="count", palette="ch:.25", data=titanic);
```

```python
# Horizontal bar plot
sns.catplot(y="deck", hue="class", kind="count",
            palette="pastel", edgecolor=".6",
            data=titanic);
```

## Plot linear regression models

```python
# Linear relationship
anscombe = sns.load_dataset("anscombe")
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           ci=None, scatter_kws={"s": 80});
```
